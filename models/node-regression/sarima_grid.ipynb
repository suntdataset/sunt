```python
# create a set of sarima configs to try
def sarima_configs(seasonal=[0]):
    models = list()
    # define config lists
    p_params = range(2)
    d_params = range(2)
    q_params = range(5)
    t_params = ['n','c','t','ct']
    P_params = range(5)
    D_params = range(2)
    Q_params = range(2)
    m_params = range(2)
    # create config instances
    for p in p_params:
        for d in d_params:
            for q in q_params:
                for t in t_params:
                    for P in P_params:
                        for D in D_params:
                            for Q in Q_params:
                                for m in m_params:
                                    cfg = [(p, d, q), (P, D, Q, m), t]
                                    models.append(cfg)
    
    return models
```
